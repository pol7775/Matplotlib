#Barplot

```python
import matplotlib.pyplot as plt

x=["python", "c", "c++", "java"]
y=[80,70,60,87]

plt.xlabel('Language', fontsize=20)
plt.ylabel('Number', fontsize=20)
plt.title('Plot', fontsize=30)
c=['y', 'r', 'b', 'g']

plt.bar(x,y, width=0.4, color=c, label='Palash')
plt.legend()
plt.show
```

```python
import matplotlib.pyplot as plt

x=["python", "c", "c++", "java"]
y=[80,70,60,87]
z=[20,50,70,80]

plt.xlabel('Language', fontsize=20)
plt.ylabel('Number', fontsize=20)
plt.title('Plot', fontsize=30)
c=['y', 'r', 'b', 'g']

plt.bar(x,y, width=0.4, color='r', label='Palash',)
plt.bar(x,z, width=0.4, color='b', label='Palash1')
plt.legend()
plt.show
```

```python
import matplotlib.pyplot as plt
import numpy as np

x=["python", "c", "c++", "java"]
y=[80,70,60,87]
z=[20,50,70,80]

width=0.2
p=np.arange(len(x))
p1=[j+width for j in p]

plt.xlabel('Language', fontsize=20)
plt.ylabel('Number', fontsize=20)
plt.title('Plot', fontsize=30)


plt.bar(p,y, width, color='r', label='Palash')
plt.bar(p1,z, width, color='b', label='Palash1')


plt.xticks(p+width/2,x,rotation=30)
plt.legend()
plt.show
```

```python
import matplotlib.pyplot as plt

x=["python", "c", "c++", "java"]
y=[80,70,60,87]
z=[20,50,70,80]

plt.xlabel('Language', fontsize=20)
plt.ylabel('Number', fontsize=20)
plt.title('Plot', fontsize=30)


plt.barh(x,y, color='r', label='Palash',)
plt.barh(x,z, color='b', label='Palash1')
plt.legend()
plt.show
```
