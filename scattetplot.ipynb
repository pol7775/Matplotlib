```python
import matplotlib.pyplot as plt

day=[1,2,3,4,5,6]
no=[1,2,3,4,7,8]

plt.scatter(day,no, color='r')
plt.title('Scatter plot',fontsize=20)
plt.xlabel('Day',fontsize=15)
plt.ylabel('No',fontsize=15)
plt.show()
```

```python
import matplotlib.pyplot as plt

day=[1,2,3,6]
no=[1,2,7,8]
colors=['r', 'b', 'g', 'm']
size=[5,9,8,10]

plt.scatter(day, no, c=colors, s=size)
plt.title('Scatter plot',fontsize=20)
plt.xlabel('Day',fontsize=15)
plt.ylabel('No',fontsize=15)
plt.show()
```
